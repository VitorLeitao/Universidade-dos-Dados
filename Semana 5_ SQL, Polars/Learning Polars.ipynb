O comando select faz basicamente varios filtros, agrupamentos, ordenamento, etc em paralelo, funcionamento muito semelhante ao SELECT do SQL

```python
# Select
df.select([
    pl.col('MonthlyCharges').filter(pl.col('MonthlyCharges')>115).sort(descending=True).head(),
])
```

```python
# Fazendo algumas analises semelhantes ao describe

# Quando vamos fazer duas analises sobre a mesma coluna, temos que renomear usando .alias('new_name')
df.select(
    pl.sum('MonthlyCharges').alias('Soma MonthlyCharges'),
    pl.min('MonthlyCharges').alias('Minimo MonthlyCharges'),
    pl.max('MonthlyCharges').alias('Maximo MonthlyCharges'),
    pl.mean('MonthlyCharges').alias('Media MonthlyCharges'),
    pl.std('MonthlyCharges').alias('Desvio padrao MonthlyCharges')
)
```

```python
# Podemos fazer analises em varias colunas dde uma vez
df.select(
    pl.col('MonthlyCharges').filter(pl.col('MonthlyCharges') > 118).count(),
    pl.col('Contract').filter(pl.col('Contract') == 'Month-to-month').count(),
    pl.col('PaperlessBilling').filter(pl.col('PaperlessBilling') == 'Yes').count()
)
```

Não necessariamente todas as consultas tem que retornar a mesma quantidade de linhas, como SQL, se por exemplo, um select retornar o maximo de uma coluna e o outro retornar as 5 primeiras linhas, para essas 5 primeiras linhas, os valores da media se repetirão

Explorando o Group by


```python
df.groupby("gender").mean()
```

Podemos tambem especificar alguma estatistica para alguma coluna especifica usando o .agg

```python
df.groupby("gender").agg(pl.col('MonthlyCharges').mean())
```

Podemos gerar mais de uma estatistica com agg

```python
df.groupby("gender").agg(
        [
            pl.count(),
            pl.col("MonthlyCharges").mean(),
            pl.col("TotalCharges").mean(),
        ]
    )
```

```python
# Vamos agrupar por dependents e pegar a media do renure(meses quie a pessoa ficou na empresa)
df.groupby('Dependents').agg(
    [
        pl.col('tenure').mean().alias('permanencia Media')
    ]
)
```
